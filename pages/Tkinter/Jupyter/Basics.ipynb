```python
from tkinter import *


root = Tk()
root.title("Data quality")
root.iconbitmap('icon.ico')

#Menu configuration
my_menu = Menu(root)
root.config(menu=my_menu)

#Create file

def file_new():
    hide_all_frames()
    root.geometry('400x400')
    file_new_frame.pack(fill="both",expand=1)
    Label(file_new_frame,text="Something apear").pack()
    
    
def edit_cut():
    hide_all_frames()
    root.geometry('400x400')
    edit_cut_frame.pack(fill="both",expand=1)    

def hide_all_frames():
    for widget in file_new_frame.winfo_children():
        widget.destroy()
    for widget in edit_cut_frame.winfo_children():
        widget.destroy()    
        
    file_new_frame.pack_forget()
    edit_cut_frame.pack_forget()
#pack_forget - to hide not to kill

    
file_menu = Menu(my_menu,tearoff=False)
my_menu.add_cascade(label="File",menu=file_menu)
file_menu.add_command(label="New...",command = file_new)
file_menu.add_separator()
file_menu.add_command(label="Exit",command=root.quit)

#Create edit
edit_menu = Menu(my_menu,tearoff=False)
my_menu.add_cascade(label="Edit",menu=edit_menu)
edit_menu.add_command(label="Cut",command = edit_cut)
edit_menu.add_separator()
edit_menu.add_command(label="Copy")

#Create frames

file_new_frame = Frame(root, width=400,height=400,bg="red")
edit_cut_frame = Frame(root, width=400,height=400,bg="blue")
root.mainloop()
```

```python
# SELECT MORE ITEMS
from tkinter import *


root = Tk()
root.title("Data quality")
root.iconbitmap('icon.ico')



#Create frame
my_frame = Frame(root)
my_scroll = Scrollbar(my_frame,orient=VERTICAL)


my_list=Listbox(my_frame,yscrollcommand=my_scroll.set,selectmode=MULTIPLE)
my_scroll.config(command=my_list.yview)
my_scroll.pack(side=RIGHT,fill=Y)
my_frame.pack()
my_list.pack(pady=15)

#Add list of items
my_list2 = [
    'One',
    'Two',
    'Three',
    'One',
    'Two',
    'Three',
    'One',
    'Two',
    'Three',
    'Two',
    'Three',
    'One',
    'Two',
    'One',
    'Two',
    'Three',
    'One',
    'Two',
    'Three',
    'One',
    'Two',
    'Three',
    'Two',
    'Three',
    'One',
    'Two'
]

for item in my_list2:
    my_list.insert(END,item)

def select_all():
    result=''
    
    for item in my_list.curselection():
        result +=str(my_list.get(item))+'\n'
    
    my_label.config(text=result)
    
    
def delete_selected():
    for item in reversed(my_list.curselection()):
        my_list.delete(item)
    
    
global my_label

my_button3 = Button(root,text='Select All',command=select_all).pack()
my_button5 = Button(root,text='Delete Selected',command=delete_selected).pack()
my_label = Label(root,text='')
my_label.pack(pady=5)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
import pandas as pd

root = tk.Tk()


style = ttk.Style()


style.map("Treeview",
         foreground=[('active','black')],
         background=[('active','white')])




sample = pd.DataFrame({'month': [1, 4, 7, 10],
                   'year': [2012, 2014, 2013, 2014],
                   'sale': [55, 40, 84, 31]})

df = pd.DataFrame(sample)
cols = list(df.columns)


tree = ttk.Treeview(root)
tree.pack()
tree["columns"] = ('month','year','sale')

tree.column("#0", width=0,stretch=NO)
for i in cols:
    tree.column(i, anchor="w",width=60)
    tree.heading(i, text=i, anchor='w')

for index, row in df.iterrows():
    if row[1]>2013:
        tree.insert("",END,text='',values=list(row),tag="falty")
    else:
        tree.insert("",END,text='',values=list(row))    

        
tree.tag_configure('falty', background='red')

root.mainloop()
```
